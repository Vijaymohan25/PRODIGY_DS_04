```python
# Train-test split for features (X) and target variable (y) in both training and testing datasets

# For the training dataset (df_train)
X_train = df_train.drop(['Sentiment'], axis=1)
y_train = df_train['Sentiment']

# For the testing dataset (df_test)
X_test = df_test.drop(['Sentiment'], axis=1)
y_test = df_test['Sentiment']
```

```python
#label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.fit_transform(y_test)
```

```python
X_train.drop(['Brand'],axis=1,inplace=True)
X_test.drop(['Brand'],axis=1,inplace=True)
```
